```python
import numpy as np
import matplotlib.pyplot as plt

#FUNCIONES ------------------------------------------------------------------------------------------------------------------------------------------------------

def generate_onda(reset, clk, vita_time, vita_time_trigger, len_PRI, duration_wave, type_wave, Fs, Bw, f_tone, T):
    if reset:
        return [], 0, None, None

    # Generar la onda según el tipo especificado
    # CHIRP ---------------------------------------------------------------------------------
    if type_wave == 0:
        t = np.linspace(0, T, int(Fs * T))
        f0 = 0
        f1 = Bw
        phase = 2 * np.pi * ((f0 + (f1 - f0) * t / (2 * T)) * t)
        I = np.cos(phase)
        Q = np.sin(phase)

        plt.figure(10)
        plt.subplot(2,1,1)
        plt.plot(I)
        plt.title('Chirp I')
        plt.xlabel('Muestras')
        plt.ylabel('Amplitud')
        plt.subplot(2,1,2)
        plt.plot(Q)
        plt.title('Chirp Q')
        plt.xlabel('Muestras')
        plt.ylabel('Amplitud')
        plt.show()

    # TONO ---------------------------------------------------------------------------------
    elif type_wave == 1:
        t = np.linspace(0, T, int(Fs * T))
        phase = 2 * np.pi * f_tone * t
        I = np.cos(phase)
        Q = np.sin(phase)

        plt.figure(11)
        plt.subplot(2,1,1)
        plt.plot(I)
        plt.title('Tono I')
        plt.xlabel('Muestras')
        plt.ylabel('Amplitud')
        plt.subplot(2,1,2)
        plt.plot(Q)
        plt.title('Tono Q')
        plt.xlabel('Muestras')
        plt.ylabel('Amplitud')
        plt.show()

    # RAMPA ---------------------------------------------------------------------------------
    elif type_wave == 2:
        t = np.linspace(0, T, int(Fs * T))
        I = np.linspace(0, 1, int(Fs * T))  # Rampa de -1 a 1
        Q = np.linspace(0, 1, int(Fs * T))  # Debe ser igual a I
        #ramp = np.mod(t, 1)
        #I = np.cos(2 * np.pi * ramp)
        #Q = np.sin(2 * np.pi * ramp)

        plt.figure(12)
        plt.subplot(2,1,1)
        plt.plot(I)
        plt.title('Ramp I')
        plt.xlabel('Muestras')
        plt.ylabel('Amplitud')
        plt.subplot(2,1,2)
        plt.plot(Q)
        plt.title('Ramp Q')
        plt.xlabel('Muestras')
        plt.ylabel('Amplitud')
        plt.show()

    else:
        raise ValueError("Tipo de onda no válido")

    # Convertir a 16 bits
    I = (I * 32767).astype(np.int16)
    Q = (Q * 32767).astype(np.int16)

    # Empaquetar en 32 bits (I en 16 bits altos, Q en 16 bits bajos)
    IQ_32 = (I.astype(np.int32) << 16) | (Q.astype(np.int32) & 0xFFFF)

    IQ_64 = np.empty((int(Fs * T)), dtype=np.int64)

    for i in range(int(Fs * T)):
        if i != (int(Fs * T) - 1):
            IQ_64[i]=((IQ_32[i].astype(np.int64) << 32) | (IQ_32[i+1].astype(np.int64) & 0xFFFFFFFF))
        else:
            IQ_64[i]=((IQ_32[i].astype(np.int64) << 32) | (0x00000000))

    o_tdata=IQ_64
    o_tvalid = 1
    return o_tdata, o_tvalid, I, Q #I y Q mandado para comprobaciones (ELIMINAR)

#--------------------------------------------------------------------------------------------------------------------------------------------------------------

# Parámetros generales
ds = 1e6  # Frecuencia de muestreo 1 MHz
bw = 100000  # Ancho de banda del chirp
f_tone = 10000  # Frecuencia del tono
T = 0.0001  # Duración de 10 ms

# Puertos del sistema
reset = 0
clk = 1
vita_time = 0
vita_time_trigger = 0
len_PRI = 512
duration_wave = 512

type_wave = 1 # 0: Chirp / 1: Tono / 2: Rampa

o_tdata, o_tvalid, I, Q = generate_onda(reset, clk, vita_time, vita_time_trigger, len_PRI, duration_wave, type_wave, ds, bw, f_tone, T)

# Extraer I (los 16 bits más altos) y Q (los 16 bits más bajos)
extracted_I = ((o_tdata >> 48) & 0xFFFF).astype(np.int16)
extracted_Q = ((o_tdata >> 32) & 0xFFFF).astype(np.int16)

print(I)
print(Q)
print(I-extracted_I)
print(Q-extracted_Q)

if o_tvalid:
    plt.figure()
    plt.plot(extracted_I, 'b', label='I extraído')
    plt.plot(extracted_Q, 'r', label='Q extraído')
    plt.title(f'Salida I/Q para tipo de onda {type_wave}')
    plt.xlabel('Muestras')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.grid()
    plt.show()
else:
    print('Error en la generación de la onda')

#extracted_I2 = ((o_tdata >> 16) & 0xFFFF).astype(np.int16)
#extracted_Q2 = ((o_tdata >> 0) & 0xFFFF).astype(np.int16)

#for i in range(int(ds * T)):

#    print(bin(extracted_I[i]),  bin(extracted_Q[i]))
#    print(bin(I[i]),  bin(Q[i]))

#    print(bin(extracted_I2[i]),  bin(extracted_Q2[i]))
#    print(bin(I[i+1]),  bin(Q[i+1]))

#    print(bin(o_tdata[i]))



```
